```python
f = open('zhangcleandata.txt', 'r')
line = f.readline()
v = []
alist = [0]
pre = 0
time = 10.0
tmp = 0.0
while line: 
#     print(line, end = '')
    lines = line.split(',')
#     print(lines)
#     print(lines)
    if len(lines) == 4:
        v2 = float(lines[2])
        a = (v2 - pre) / (time + tmp)
        pre = v2
        tmp = 10.0
        alist.append(a)
    else:
        if lines[0] == "XXXXXXXXX\n":
            tmp += 10.0
    line = f.readline() 
f.close() 
# print(alist)

x = range(len(alist))
y = alist
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
# plt.ylim(-1, 1)  # 限定纵轴的范围
plt.plot(x, y)
plt.rcParams['axes.unicode_minus']=False
# plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"AIS序列") #X轴标签
plt.ylabel("加速度(m/s2)") #Y轴标签
plt.title("") #标题
plt.savefig("a_clean_zhang.png")
plt.show()
```

```python
f = open('zhangcleandata.txt', 'r')
line = f.readline()
cog = []
while line: 
#     print(line, end = '')
    lines = line.split(',')
#     print(lines)
#     print(lines)
    if len(lines) == 4:
        c = lines[3].split('\n')
        cog.append(float(c[0]))
    line = f.readline() 
f.close() 

x = range(len(cog))
y = cog
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
# plt.ylim(-1, 1)  # 限定纵轴的范围
plt.plot(x, y)
plt.rcParams['axes.unicode_minus']=False
# plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"AIS序列") #X轴标签
plt.ylabel("航向（°）") #Y轴标签
plt.title("") #标题
plt.savefig("cog_clean_zhang.png")
plt.show()
```

```python
f = open('zhangcleandata.txt', 'r')
line = f.readline()
v = []
rate = [0]
pre = 350
time = 10.0
tmp = 0.0
while line: 
#     print(line, end = '')
    lines = line.split(',')
#     print(lines)
#     print(lines)
    if len(lines) == 4:
        c = float(lines[3].split('\n')[0])
        ra = (c - pre) / (time + tmp)
        pre = c
        tmp = 10.0
        if ra < 0:
            ra = ra * -1
        rate.append(ra)
    else:
        if lines[0] == "XXXXXXXXX\n":
            tmp += 10.0
    line = f.readline() 
f.close() 
# print(alist)

x = range(len(rate))
y = rate
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
# plt.ylim(-1, 1)  # 限定纵轴的范围
plt.plot(x, y)
plt.rcParams['axes.unicode_minus']=False
# plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"AIS序列") #X轴标签
plt.ylabel("转向速率(°/s)") #Y轴标签
plt.title("") #标题
plt.savefig("rate_clean_zhang.png")
plt.show()
```
