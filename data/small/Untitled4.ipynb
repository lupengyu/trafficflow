```python
f = open('clean_repair-trajectory_new.txt', 'r')
line = f.readline()
v = []
rate = []
pre = 318
pretime = -817407
while line: 
#     print(line, end = '')
    lines = line.split(',')
#     print(lines)
#     print(lines)
    if len(lines) == 5:
        time = float(lines[4].split('\n')[0])
#         print(time, pretime)
        if time == pretime:
            rate.append(0)
        else:
            c = float(lines[3])
            tmp = c - pre
            if tmp < 0:
                tmp = tmp * -1
            up = tmp
            if tmp > 180:
                up = 360 - tmp
            ra = up / (time - pretime)
            pre = c
            if ra < 0:
                ra = ra * -1
            rate.append(ra)
    line = f.readline() 
f.close() 
# print(alist)
print(rate[0:10])

x = range(len(rate))
y = rate
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
# plt.ylim(-1, 1)  # 限定纵轴的范围
plt.plot(x, y)
plt.rcParams['axes.unicode_minus']=False
# plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"AIS序列") #X轴标签
plt.ylabel("转向速率(°/s)") #Y轴标签
plt.title("") #标题
plt.savefig("rate_clean_repair-trajectory_new_2.png")
plt.show()
```

```python
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']
f = open('clean_repair-trajectory_new.txt', 'r')
line = f.readline()
v = []
time = []
while line: 
#     print(line, end = '')
    lines = line.split(',')
#     print(lines)
#     print(lines)
    if len(lines) == 5:
        t = float(lines[4].split('\n')[0])
        time.append(t)
        v.append(float(lines[2]))
    line = f.readline() 
f.close() 

tmp = time[100]
for i in range(len(time)):
    time[i] = (time[i] - tmp) / 3600

names = ['5', '10', '15', '20', '25']
x = time[100:200]
y = v[100:200]
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
# plt.ylim(-1, 110)  # 限定纵轴的范围
plt.plot(x, y)
# plt.legend()  # 让图例生效
# plt.xticks(x, names, rotation=45)
# plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"时间差(h)") #X轴标签
plt.ylabel("速度(nmi/h)") #Y轴标签
plt.title("") #标题
plt.savefig("a_clean_repair-trajectory_new_100_200.png")
plt.show()

# fig.tight_layout()
```
